```python
import pandas as pd
from sqlalchemy import create_engine, func
from  db_conn import url
import psycopg2
```

```python
postgresInfo = f'postgresql://{url}'
conn_string = postgresInfo

db = create_engine(conn_string)
conn = db.connect()

conn = psycopg2.connect(conn_string)
conn.autocommit = True
cursor = conn.cursor()
```

```python
concussion_start = pd.read_sql_query('''SELECT * FROM clean_concussion_data;''', db)
```

```python
concussion_start.dtypes
```

```python
concussion_start["pkey"] = (concussion_start["Season"].astype(str) + concussion_start["Player"])
concussion_start.head()
```

```python
concussion_final = concussion_start[["pkey", "Season", "Player", "Team", "Position", "Games_Missed"]].copy()
```

```python
concussion_final.head()
```

```python
# Write the dataframe to postgres database
concussion_final.to_sql('concussion_final', db, if_exists='replace',
          index=False)
sql1 = '''select * from clean_concussion_data;'''
cursor.execute(sql1)
for i in cursor.fetchall():
    print(i)
  
conn.commit()
conn.close()
```
