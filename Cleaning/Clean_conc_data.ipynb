```python
import pandas as pd
```

```python
df = pd.read_csv("NHL Concussions Database_data.csv")
df.head()
```

```python
clean_df = df.drop(['Injury Type', 'Cap Hit', 'Chip', 'Games Missed.1'], axis = 1)
clean_df.head(20)
```

```python
clean_df['Player'] = clean_df['Player'].apply(lambda x: ' '.join(x.split(', ')[::-1]))
clean_df.head()
```

```python
clean_df['Position'].value_counts()
```

```python
clean_df['Team'].value_counts()
```

```python
clean_df = clean_df[clean_df['Position'] != 'G']
```

```python
clean_df['Position'].value_counts()
```

```python
clean_df['Season'] = clean_df['Season'].str[:4]
clean_df.head()
```

```python
city_dict = {'Anaheim': 'ANA', 'Arizona': 'ARI', 'Boston': 'BOS', 'Buffalo': 'BUF', 'Carolina': 'CAR', 'Columbus': 'CBJ', 'Calgary': 'CGY', 'Chicago': 'CHI', 'Colorado': 'COL', 'Dallas': 'DAL',
             'Detroit': 'DET', 'Edmonton': 'EDM', 'Florida': 'FLA', 'Los Angeles': 'L.A', 'Minnesota': 'MIN', 'Montreal': 'MTL', 'New Jersey': 'N.J', 'Nashville': 'NSH', 'NY Islanders': 'NYI', 'NY Rangers': 'NYR',
             'Ottawa': 'OTT', 'Philadelphia': 'PHI', 'Phoenix': 'ARI', 'Pittsburgh': 'PIT', 'St. Louis': 'STL', 'San Jose': 'S.J', 'Tampa Bay': 'T.B', 'Toronto': 'TOR', 'Vancouver': 'VAN', 'Vegas': 'VGK',
             'Winnipeg': 'WPG', 'Washington': 'WSH',}
```

```python
clean_df['Team'] = clean_df['Team'].replace(city_dict)
clean_df.head()
```

```python
clean_df['Team'].value_counts()
```

```python
clean_df = clean_df.rename(columns = {'Games Missed': 'Games_Missed'})
clean_df.head()
```

```python
clean_df.to_csv('clean_concussion_data.csv')
```
