```python
moneypuck_master_df = pd.concat([moneypuck2012_df, moneypuck2013_df, moneypuck2014_df,
                                 moneypuck2015_df, moneypuck2016_df, moneypuck2017_df,
                                 moneypuck2018_df, moneypuck2019_df, moneypuck2020_df,
                                 moneypuck2021_df, moneypuck2022_df], axis = 0)

moneypuck_master_df
```

```python
moneypuck_master_clean_rows = moneypuck_master_df.drop(
    moneypuck_master_df[(moneypuck_master_df.situation == "other") |
                        (moneypuck_master_df.situation == "5on5") |
                        (moneypuck_master_df.situation == "5on4") |
                        (moneypuck_master_df.situation == "4on5")].index)
```

```python
moneypuck_master_clean_rows
```

```python
moneypuck_master_clean_rows.to_csv("../Resources/moneypuck_rows.csv")

list(moneypuck_master_clean_rows.columns.values)
```

```python
moneypuck_cleaned = moneypuck_master_clean_rows[['season', 'name', 'team', 'position', 'games_played','onice_corsipercentage', 'onice_fenwickpercentage', 'i_f_goals', 'i_f_points', 'i_f_penalityminutes','i_f_hits', 'i_f_takeaways', 'i_f_giveaways']].copy()
```

```python
moneypuck_cleaned = moneypuck_cleaned.rename(columns ={'onice_corsipercentage':'onIce_corsiPercentage', 'onice_fenwickpercentage':'onIce_fenwickPercentage','i_f_goals':'I_F_goals', 'i_f_points':'I_F_points', 'i_f_penalityminutes':'I_F_penalityMinutes','i_f_hits':'I_F_hits', 'i_f_takeaways':'I_F_takeaways','i_f_giveaways':'I_F_giveaways'})
```

```python
# Write the dataframe to postgres database
moneypuck_cleaned.to_sql('moneypuck_cleaned', db, if_exists='replace',
          index=False)
sql1 = '''select * from moneypuck_cleaned;'''
cursor.execute(sql1)
for i in cursor.fetchall():
    print(i)
  
conn.commit()
conn.close()
```
